# Git 

	I. SCM (Software Configuration Management) : 소프트웨어 형상 관리
	II. Git                                
		1. 설치  -http://git-scm.com
			- Book : Pro Git (https://git-scm.com/book/ko/v2) 
		2. 사용법
			- bash
			- gui
		3. 버전관리 용어 
			1) 알파버전(Alpha) : 개발자들이 테스트하는 버전
			2) 베타버전(Beta)   : 사용자들이 테스트하는 버전
			3) Milestone : 1.0-M1,1.0-M2, …
			4) RC(Realease Candidate) : 1.0-RC1,1.0-RC2, …
			5) GA(General Availability : 1.0, 1.0.GA, …           > 정식 배포
		4. 버전관리 전략(subversion 기준)
			1) branch
			2) tag
			3) trunk - 개발중 (거의 완성)
		5. 실습
			1) GIT을 설치하고 가장 먼저 해야할 일
				- git config --global user.name"lty516"
				- git config --global user.email lty516@naver.com
				
				- git config --list   // git 정보보기
				- git config user.name
			2) 프로젝트 생성 후 가장 먼저 해야할 일
				- 저장소(Local Repository) 생성 
					git init
					git clone <github주소>
				- 2단계 작업(WD ->Stage)
					git add first.txt
					git add 파일명, 파일명, …
					git add *.txt
					git add --all 또는 git add .
					
					git status  // 상태확인
				- 3단계 작업(Stage -> local repository) :commit
					git commit
					> vi에디터 I로 메시지 작성 > esc > : wq
					git commit -m "메세지" > 메시지 바로 작성
			3) Remote Repository
				- 직접 서버 구축(artifactory)
				- Web으로 구현된 서버 : github, gitlab, bitbucket, CodeBase, CloudForge, …

				- 명령어
					® git remote, git remote -v	
					git remote add <github주소> 	
					git push <업로드할 서버주소> <올릴 데이터가 있는 저장소 이름(로컬측)> 	
					git pull <다운로드 받을 서버주소> <받을 데이터가 있는 저장소 이름(서버측)>	fetch + merge
					git fetch <다운로드 받을 서버주소> 	
					git merge <받을 데이터가 있는 저장소 이름(서버측)> 	
			4) commit 되돌리기 : reset , revert
					® git reset --[hard | soft | mixed]	
					git reset --hard <checksum>	
					git push origin +master	강제 업데이트	
					git push -fu origin master
					
			5) branch
					git branch login
					
					git branch    : login, master 두개가 나오늘 것이 보인다. 초록색 master - 작업중이다.
					
					git checkout login   : 헤드를 옮길때 (login으로 옮김)
					
					git branch    : login이 초록
					
				
			6) 협동
					- collaboration
							팀장과 팀원이 한 조로 구성(팀장 1명)
							팀장
								해당 프로젝트의 settings로 이동
								Manage Access선택
							팀원 
								프로젝트를 clone으로 생성
								브랜치 생성 후 작업
								서버로 push
					- fork
							https://github.com/octocat/Spoon-Knife
